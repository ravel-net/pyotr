```python
%run startup.py
bf = Session(host="localhost")
```

```python
# Assign a friendly name to your network and snapshot
NETWORK_NAME = "T1"
SNAPSHOT_NAME = "T1_no_backup_links"

SNAPSHOT_PATH = "./networks/t1/no_backup_links"

# Now create the network and initialize the snapshot
bf.set_network(NETWORK_NAME)
bf.init_snapshot(SNAPSHOT_PATH, name=SNAPSHOT_NAME, overwrite=True)
```

```python
interface_properties = bf.q.interfaceProperties().answer().frame()
interface_properties
```

```python
# Do a traceroute from host1 to 1.0.2.2
tr_frame = bf.q.traceroute(startLocation="source", headers=HeaderConstraints(dstIps="10.0.5.20")).answer().frame()

# Display results using customizations to handle large string values

show(tr_frame)
```

```python
# Assign a friendly name to your network and snapshot
BASE_NETWORK_NAME = "T1"
BASE_SNAPSHOT_NAME = "T1_whole_topo"

BASE_SNAPSHOT_PATH = "./networks/t1/whole_topo"

# Now create the network and initialize the snapshot
bf.set_network(BASE_NETWORK_NAME)
bf.init_snapshot(BASE_SNAPSHOT_PATH, name=BASE_SNAPSHOT_NAME, overwrite=True)
```

```python
node_properties = bf.q.nodeProperties().answer().frame()
node_properties.columns
```

```python
interface_properties = bf.q.interfaceProperties().answer().frame()
interface_properties[["Interface", "Active",  "All_Prefixes", "Incoming_Filter_Name", "Outgoing_Filter_Name", "Primary_Address", "Primary_Network"]]
```

```python
# Fork a new snapshot with backup links inactive
BLINKS_INACTIVE_SNAPSHOT_NAME = "two_b_links_inactive"

interfaces = [
    Interface(hostname="r_1", interface="Ethernet2"),
    Interface(hostname="r_2", interface="Ethernet3")
    ]
bf.fork_snapshot(BASE_SNAPSHOT_NAME, BLINKS_INACTIVE_SNAPSHOT_NAME, deactivate_interfaces=interfaces, overwrite=True)
```

```python
interface_properties = bf.q.interfaceProperties().answer(BLINKS_INACTIVE_SNAPSHOT_NAME).frame()
interface_properties[["Interface", "Active",  "All_Prefixes", "Incoming_Filter_Name", "Outgoing_Filter_Name", "Primary_Address", "Primary_Network"]]
```

```python
# Do a traceroute from host1 to 1.0.2.2
tr_frame = bf.q.traceroute(startLocation="source1", headers=HeaderConstraints(dstIps="r_w")).answer().frame()

# Display results using customizations to handle large string values

show(tr_frame)
```

```python
aclAns = bf.q.filterLineReachability(nodes="r_2").answer()
show(aclAns.frame())
```

```python
result = bf.q.searchFilters(headers=HeaderConstraints(srcIps='0.0.0.0/24', dstIps='dest'), filters='ACL2').answer().frame()
result
```

```python
# Check if a representative host can reach the DNS server
dns_flow = HeaderConstraints(srcIps="10.0.2.20/24",
                             dstIps="10.0.2.2/24")
answer = bf.q.testFilters(headers=dns_flow,
                         nodes="r_2",
                         filters="ACL2").answer()
show(answer.frame())
```
